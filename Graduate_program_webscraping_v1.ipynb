```python
########all combined ####


import urllib
import requests
import json
import pandas as pd


##put the last page no. from range 1 to n....here last page no. is 12

for i in range(1, 12):
    url = "https://www.graduateprograms.com/chimera/v1/school-review-list?mainTemplate=school-review-list&path=%2Fschools%2Fbrown-university&isDataTarget=false&schoolId=3744&programId=&level=&page="+str(i)+"&perPage=10&truncationLength=0&readMoreOmission=...&readMoreText=Read%20More"
    data = urllib.request.urlopen(url).read()
    theJSON = json.loads(data)
    res = {key: theJSON[key] for key in theJSON.keys() 
                               & {'content'}} 
    res_list=list(res.values())
    new_list = [i["reviews"] for i in res_list]
    index = 0
    final_list = new_list[index]
    
    df = pd.DataFrame(final_list)
    print(df)
    df.to_excel(r'C:\Users\prerana.a.singh\OneDrive - Accenture\CMRI\Graduateprogram\truthuniversity\data' + str(i) + '.xlsx', index=False)
    
    
from glob import glob
file_list = glob(r'C:\Users\prerana.a.singh\OneDrive - Accenture\CMRI\Graduateprogram\truthuniversity\data' + '*.xlsx')
df_list = [pd.read_excel(file) for file in file_list]
complete_df = pd.concat(df_list, axis=0).drop_duplicates() 
complete_df.to_excel(r'C:\Users\prerana.a.singh\OneDrive - Accenture\CMRI\Graduateprogram\truthuniversity\data\combined.xlsx', index=False)    
```
